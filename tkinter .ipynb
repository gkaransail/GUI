```python
import tkinter as tk
from tkinter import messagebox
import openpyxl
from openpyxl import Workbook

# Function to save form data into an Excel file
def save_to_excel(name, email, age, occupation):
    file_name = "form_output.xlsx"
    
    # Create a new workbook and add data
    try:
        workbook = openpyxl.load_workbook(file_name)
        sheet = workbook.active
    except FileNotFoundError:
        workbook = Workbook()
        sheet = workbook.active
        # Create headers for new sheet
        sheet.append(["Name", "Email", "Age", "Occupation"])

    # Add form data
    sheet.append([name, email, age, occupation])
    
    # Save the workbook
    workbook.save(file_name)
    messagebox.showinfo("Success", f"Data saved to {file_name}")

# Function to be called when the form is submitted
def submit_form():
    name = name_entry.get()
    email = email_entry.get()
    age = age_entry.get()
    occupation = occupation_entry.get()
    
    if not name or not email or not age or not occupation:
        messagebox.showwarning("Input Error", "Please fill all the fields!")
    else:
        save_to_excel(name, email, age, occupation)
        clear_form()

# Function to clear the form after submission
def clear_form():
    name_entry.delete(0, tk.END)
    email_entry.delete(0, tk.END)
    age_entry.delete(0, tk.END)
    occupation_entry.delete(0, tk.END)

# Set up the GUI window
root = tk.Tk()
root.title("User Information Form")

# Create form labels and entries
tk.Label(root, text="Name").grid(row=0, column=0, padx=10, pady=10)
name_entry = tk.Entry(root)
name_entry.grid(row=0, column=1, padx=10, pady=10)

tk.Label(root, text="Email").grid(row=1, column=0, padx=10, pady=10)
email_entry = tk.Entry(root)
email_entry.grid(row=1, column=1, padx=10, pady=10)

tk.Label(root, text="Age").grid(row=2, column=0, padx=10, pady=10)
age_entry = tk.Entry(root)
age_entry.grid(row=2, column=1, padx=10, pady=10)

tk.Label(root, text="Occupation").grid(row=3, column=0, padx=10, pady=10)
occupation_entry = tk.Entry(root)
occupation_entry.grid(row=3, column=1, padx=10, pady=10)

# Submit button
submit_btn = tk.Button(root, text="Submit", command=submit_form)
submit_btn.grid(row=4, columnspan=2, pady=10)

# Run the Tkinter event loop
root.mainloop()
```
